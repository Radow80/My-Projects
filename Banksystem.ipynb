```python
def createfile(name):
    # Function to create a file if it doesn't exist or open it for appending
    try:
        with open(name, "a+") as new:  # Open the file in append mode; creates file if it doesn't exist
            new.write("")  # Ensure file is accessible and writable
        print(f"File '{name}' has been created successfully.")
    except Exception as e:
        print(f"Failed to create file '{name}'. Error: {e}")


class Account:
    # Represents a single bank account
    def __init__(self, name, address, balance=0):
        # Initialize account with name, address, and balance
        self.name = name
        self.address = address
        self.balance = balance

    def deposit(self, amount):
        # Add a positive amount to the balance
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount

    def withdraw(self, amount):
        # Deduct a valid amount from the balance
        if amount <= 0 or amount > self.balance:
            raise ValueError("Invalid withdrawal amount.")
        self.balance -= amount

    def __str__(self):
        # Return a string representation of the account
        return f"{self.name} : {self.address} : {self.balance}"


class Bank:
    # Represents a collection of accounts managed in a file
    Main_file = "FCBank.txt"  # Default file name for storing account data

    def __init__(self, file=Main_file):
        # Initialize the bank and load accounts from the file
        createfile(file)
        self.allaccounts = {}  # Dictionary to store account data
        self.load_bankfile(file)

    def load_bankfile(self, file=Main_file):
        # Load accounts from the file into the allaccounts dictionary
        try:
            with open(file, "r") as f:
                for line in f.readlines():
                    # Parse each line as name, address, and balance
                    info = line.strip().split(":")
                    self.allaccounts[info[0].strip()] = {
                        "Address": info[1].strip(),
                        "Balance": float(info[2].strip()),
                    }
        except Exception as e:
            print(f"Error loading bank file '{file}': {e}")

    def save_accounts(self):
        # Save all accounts to the file
        with open(Bank.Main_file, "w") as f:
            for name, data in self.allaccounts.items():
                f.write(f"{name} : {data['Address']} : {data['Balance']}\n")

    def change_file(self, fname):
        # Change the file used for storing accounts
        createfile(fname)
        self.save_accounts()
        self.allaccounts = {}  # Clear the current accounts
        Bank.Main_file = fname
        self.load_bankfile(fname)

    def add_accounts(self, accounts):
        # Add multiple accounts to the bank
        for account in accounts:
            if not isinstance(account, Account):
                print(f"Invalid account: {account}")
                continue
            self.allaccounts[account.name] = {
                "Address": account.address,
                "Balance": account.balance,
            }
        self.save_accounts()

    def remove_accounts(self, names):
        # Remove accounts by name
        for name in names:
            if name in self.allaccounts:
                del self.allaccounts[name]
                print("Accounts removed successfully.")
            else:
                print(f"Account '{name}' not found.")
        self.save_accounts()

    def search_account(self, account_name):
        # Search for an account by name
        return self.allaccounts.get(account_name, None)


def sort_accounts(bank, method="ascending", key="name"):
    # Sort the accounts based on the given key and order
    if key == "name":
        reverse = method.lower() == "descending"
        return dict(sorted(bank.allaccounts.items(), reverse=reverse))
    elif key in ["Address", "Balance"]:
        reverse = method.lower() == "descending"
        return dict(sorted(bank.allaccounts.items(), key=lambda x: x[1][key], reverse=reverse))
    else:
        print("Invalid sorting key. Use 'name', 'Address', or 'Balance'.")
        return None

# Error message for invalid input
invalid = "Invalid Input: Enter correct values only (i.e digits for amount)"


def main():
    # Main function to provide an interactive menu for the bank system
    print("Welcome to the Bank System!")
    bank = Bank()  # Create a bank instance

    while True:
        # Display menu options
        print("\nMenu:")
        print("1. Add Account")
        print("2. Remove Account")
        print("3. Search Account")
        print("4. Deposit")
        print("5. Withdraw")
        print("6. Sort Accounts")
        print("7. Change File")
        print("8. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            # Add a new account
            name = input("Enter account name: ")
            address = input("Enter account address: ")
            while True:
                try:
                    balance = float(input("Enter initial balance: "))
                    break
                except:
                    print(invalid)
            account = Account(name, address, balance)
            bank.add_accounts([account])
            print("Account added successfully.")

        elif choice == "2":
            # Remove accounts
            names = input("Enter account names to remove (space-separated): ").split(" ")
            names = [name.strip() for name in names]
            bank.remove_accounts(names)
            

        elif choice == "3":
            # Search for an account
            name = input("Enter the account name to search: ")
            account = bank.search_account(name)
            if account:
                print(f"Account found: {name}, Address: {account['Address']}, Balance: {account['Balance']}")
            else:
                print("Account not found.")

        elif choice == "4":
            # Deposit amount into an account
            name = input("Enter the account name: ")
            account_info = bank.search_account(name)
            if account_info:
                while True:
                    try:
                        amount = float(input("Enter deposit amount: "))
                        break
                    except:
                        print(invalid)
                account = Account(name, account_info["Address"], account_info["Balance"])
                try:
                    account.deposit(amount)
                    bank.add_accounts([account])
                    print(f"Deposit successful. Account with name {name} now has {account.balance} in Balance.")
                except ValueError as e:
                    print(e)
            else:
                print("Account not found.")

        elif choice == "5":
            # Withdraw amount from an account
            name = input("Enter the account name: ")
            account_info = bank.search_account(name)
            if account_info:
                while True:
                    try:
                        amount = float(input("Enter withdrawal amount: "))
                        break
                    except:
                        print(invalid)
                account = Account(name, account_info["Address"], account_info["Balance"])
                try:
                    account.withdraw(amount)
                    bank.add_accounts([account])
                    print("Withdrawal successful.")
                except ValueError as e:
                    print(e)
            else:
                print("Account not found.")

        elif choice == "6":
            # Sort and display accounts
            method, key = "", ""
            while method not in ["ascending", "descending"]:
                method = input("Enter sorting method (ascending/descending) ONLY: ")
            while key not in ["Name", "Address", "Balance"]:
                key = input("Enter sorting key (name/Address/Balance) ONLY: ")
            sorted_accounts = sort_accounts(bank, method, key)
            if sorted_accounts:
                print("Sorted accounts:")
                for name, info in sorted_accounts.items():
                    print(f"{name} : {info['Address']} : {info['Balance']}")

        elif choice == "7":
            # Change the storage file
            fname = input("Enter the new file name: ")
            bank.change_file(fname)
            print(f"File changed to {fname}.")

        elif choice == "8":
            # Exit the program
            print("Exiting the system. Goodbye!")
            break

        else:
            print("Invalid choice. Please Enter digits from 1-8 as your choice.")


if __name__ == "__main__":
    main()
```
